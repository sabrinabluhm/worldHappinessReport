## World Happiness Report

### Merging the tables from the years 2005 - 2009


```python
import pandas as pd
import numpy as np

table2015 = pd.read_csv('WorldHappinessScore_2015.csv', delimiter = ';')
table2015["year"] = 2015
#table2015.head(30)

table2016 = pd.read_csv('WorldHappinessScore_2016.csv', delimiter = ';')
table2016["year"] = 2016
#table2016.head(30)

table2017 = pd.read_csv('WorldHappinessScore_2017.csv', delimiter = ';')
table2017["year"] = 2017
#table2017.head(30)

table2018 = pd.read_csv('WorldHappinessScore_2018.csv', delimiter = ';')
table2018["year"] = 2018
#table2018.head(30)

table2019 = pd.read_csv('WorldHappinessScore_2019.csv', delimiter = ';')
table2019["year"] = 2019
#table2019.head(30)

tables = [table2015, table2016, table2017, table2018, table2019]

#mergedTables = pd.concat(tables, axis=0, join='outer', ignore_index=False, keys=None,
          #levels=None, names=None, verify_integrity=False, copy=True)
mergedTables = pd.concat(tables)
```
